```python
# FILEPATH

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
import numpy as np

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('./informer_yields.csv')

# Extract the relevant columns containing the data for which we want to calculate pairwise distance
data = df / 100

# Calculate the pairwise distance using the pdist function from the scipy.spatial.distance module
pairwise_dist = pdist(data)

# Convert the pairwise distance array into a square matrix using the squareform function from the same module
square_matrix = squareform(pairwise_dist)

# Create a heatmap using seaborn's heatmap function, passing in the square matrix as the data
sns.heatmap(square_matrix)

# Show the plot
plt.show()
```

```python
a = 1.2
fig, ax = plt.subplots(figsize=(10*a, 6*a))
matrix = np.triu(np.ones_like(square_matrix, dtype=bool))
np.fill_diagonal(matrix, False)
sns.heatmap(square_matrix, annot=True, mask=matrix, fmt='.2f', ax=ax, cmap='GnBu', xticklabels=False, yticklabels="auto",
            cbar_kws={'shrink': 0.6, 'location': 'right', 'anchor': (-1.5, 0.75), 'format': "{x:.2f}", 'spacing': 'proportional', 'extend': 'both'})
# sns.heatmap(dist_mat_yield, annot=True, mask=matrix, fmt='.2f', ax=ax, cmap='GnBu', xticklabels=False, yticklabels="auto",
#             cbar_kws={'shrink': 0.6, 'location': 'right', 'anchor': (-1.5, 0.75), 'format': "{x:.2f}", 'spacing': 'proportional', 'extend': 'both'}, vmin=0, vmax=7)
# set the range of color bar in seaborn
plt.title("Euclidean distance on yields between informer yields")
# plt.show()
plt.savefig('yield_distance_informer.png', dpi=500, bbox_inches='tight', transparent=True)
```
